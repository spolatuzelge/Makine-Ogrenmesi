```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
data = pd.read_csv('data.csv')
data=data.iloc[:,1:]
data.head()
```

```python
X = data.drop('Sales', axis=1)
y = data['Sales']
```

```python
y.head()
```

```python
X.head()
```

```python
# Statsmodels
modelStas = sm.OLS(y, X).fit()
```

```python
modelStas.summary()
```

```python
modelScikit = LinearRegression()
modelScikit.fit(X, y)
```

```python
modelScikit.coef_
```

```python
modelScikit.intercept_
```

## MODEL TUNÄ°NG

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
modelScikit.fit(X_train, y_train)
```

```python
# EÄitim HatasÄ±
RMSE = mean_squared_error(y_train, modelScikit.predict(X_train))**0.5
RMSE
```

```python
# Test HatasÄ±
RMSE = mean_squared_error(y_test, modelScikit.predict(X_test))**0.5
RMSE
```

```python
# K-KatlÄ± CV :
cross_val_score(modelScikit, X, y, cv=10, scoring='neg_mean_squared_error')
```

```python
np.mean(-cross_val_score(modelScikit, X, y, cv=10, scoring='neg_mean_squared_error'))**0.5
```
