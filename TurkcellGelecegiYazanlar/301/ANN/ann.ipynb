```python
# Yapay sinir ağları

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,scale
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
```

```python
data = pd.read_csv('data.csv')
data = data.dropna()
dms = pd.get_dummies(data[['League', 'Division', 'NewLeague']])
y = data["Salary"]
X_ = data.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train) # X_train verisini standartlaştır
X_trainScale = scaler.transform(X_train) # X_train verisini standartlaştır
X_testScale = scaler.transform(X_test) # X_test verisini standartlaştır
```

```python
mlp_Model = MLPRegressor().fit(X_trainScale, y_train)
mlp_Model
```

```python
mlp_Model.predict(X_testScale)[0:5]
```

```python
y_pred = mlp_Model.predict(X_testScale)
np.sqrt(mean_squared_error(y_test, y_pred))
```

### MODEL TUNİNG

```python
mlp_params = {"alpha": [0.1, 0.01, 0.03, 0.005, 0.0001],
            "hidden_layer_sizes": [(10,20), (5,5), (100,100),]}
```

```python
mlp_cv_model = GridSearchCV(mlp_Model, mlp_params, cv=10, verbose=2, n_jobs=-1).fit(X_trainScale, y_train)
```

```python
mlp_cv_model.best_params_
```

```python
mlp_tuned = MLPRegressor(alpha=0.005, hidden_layer_sizes=(100, 100)).fit(X_trainScale, y_train)
```

```python
y_pred = mlp_tuned.predict(X_testScale)
np.sqrt(mean_squared_error(y_test, y_pred))
```
